## 2D Line Plot

```python
# Plotting a simple function
price = [48000, 54000, 57000, 49000, 47000, 45000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
# From a pandas dataframe
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# Plotting multiple plots
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# Labels title
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kholi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# Colors(hex) and line(width and style) and marker(size)
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'], color='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black')

plt.title('Rohit Sharma Vs Virat Kholi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# Multiple types of line graph
batsman = pd.read_csv('Data/sharma-kohli.csv')

# dashed
# plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed')
# plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashed')

# dotted
# plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dotted')
# plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dotted')

# dashed dotted
# plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashdot')
# plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot')

# linewidth
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='solid', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashed', linewidth=2)

plt.title('Rohit Sharma Vs Virat Kholi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# Marker and Markersize
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='solid', linewidth=3, marker='D', markersize=10)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashed', linewidth=2, marker='D')

plt.title('Rohit Sharma Vs Virat Kholi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')
```

```python
# legend -> location
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='solid', linewidth=3, marker='D', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashed', linewidth=2, marker='D', label='Rohit')

plt.title('Rohit Sharma Vs Virat Kholi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')

plt.legend()
```

```python
# Limiting axes
price = [48000, 54000, 57000, 49000, 47000, 450000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grid
batsman = pd.read_csv('Data/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='solid', linewidth=3, marker='D', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashed', linewidth=2, marker='D', label='Rohit')

plt.title('Rohit Sharma Vs Virat Kholi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Scored')

plt.grid()
```

```python
# plt.scatter on pandas data
df = pd.read_csv('Data/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'], color='red', marker='+')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

## BAR CHART

```python
# Simple Bar Chart
Children = [10, 20, 40, 10, 30]
Color = ['Red', 'Blue', 'Green', 'Yellow', 'Pink']

plt.bar(Color, Children, color='red')
plt.title('Favourite Color')
plt.xlabel('Color')
plt.ylabel('Children')
```

```python
# Horizontal Bar Chart
df = pd.read_csv('Data/batter.csv')
df = df.head(10)
df
```

```python
plt.barh(df['batter'], df['runs'])
plt.title('Runs Comparison')
plt.xlabel('Batsman')
plt.ylabel('Runs')
```

```python
# Color and label
df = pd.read_csv('Data/batsman_season_record.csv')
df
```

```python
# Multiple Bar Charts
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Using xticks
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar Chart
df = pd.read_csv('Data/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

## Pie Chart

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# using with dataset
df = pd.read_csv('Data/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'])

plt.show()
```

```python
# Percentage and color
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors=['blue','green','yellow','pink','cyan','brown'])

plt.show()
```

```python
# Explode shadow
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.1,0,0,0,0,0.1], shadow=True)

plt.show()
```
