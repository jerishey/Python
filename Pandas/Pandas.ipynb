### Filtering Rows with conditions

```python
import pandas as pd

data = {
    "Name":['Rakesh', 'Rohan', 'Sourav', 'Nitish', 'Rohit', 'Arun', 'Poonam', 'Aarti'],
    "Age":[21, 24, 21, 23, 20, 21, 21, 22],
    "Salary":[50000, 60000, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performace Score":[85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)

# display the dataframe
# print("Sample DataFrame")
# print(df)

high_salary = df[df['Salary'] > 50000]
print('Employees with salary above 50000')
print(high_salary)

# Filtering rows salary > 50000 & age > 30
filtered = df[(df['Age'] > 21) & (df['Salary'] > 50000)]
print(f'Employee list Age > 21 + Salary > 50000')
print(filtered)

# Using OR Condition
filtered_or = df[(df['Age'] > 22) | (df['Performace Score'] > 90)]
print('Employees older than 22 OR Performance score above 90')
print(filtered_or)
```

### Handling Missing Data

```python
import pandas as pd

data = {
    "Name":['Rakesh', None, 'Sourav', 'Nitish', 'Rohit', 'Arun', 'Poonam', 'Aarti'],
    "Age":[21, None, 21, 23, 20, 21, 21, 22],
    "Salary":[50000, None, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performace Score":[85, None, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(df)

# Using isnull() to find missing values in DataFrame
# True -> Data is missing & False -> Data is present
print("Handling Missing Data")
print(df.isnull())

# Using isnull() with .sum() to find total number of missing Values
print(df.isnull().sum())
```

### Using Interpolate Method for missing data

```python
import pandas as pd

data = {
    "Name":['Rakesh', 'Rohan', 'Sourav', 'Nitish', 'Rohit', 'Arun', 'Poonam', 'Aarti'],
    "Age":[21, None, 21, 23, 20, 21, 21, 22],
    "Salary":[50000, None, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performance Score":[85, None, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(df)

# The interpolate() method in Pandas is used to fill missing values (NaN) by estimating values between existing data points.
# DataFrame.interpolate(method='linear', axis=0)
df['Age'] = df['Age'].interpolate(method='linear')
print("After interpolating")
print(df)
```

## Sorting & Aggregation

```python
import pandas as pd

data = {
    "Name":['Rakesh', 'Rohan', 'Sourav', 'Nitish', 'Rohit', 'Arun', 'Poonam', 'Aarti'],
    "Age":[21, 24, 21, 23, 20, 21, 21, 22],
    "Salary":[50000, 60000, 45000, 52000, 49000, 70000, 48000, 58000],
    "Performace Score":[85, 90, 78, 92, 88, 95, 80, 89]
}

df = pd.DataFrame(data)
print(df)

# Sorting data 
print("Sorted Data")
df.sort_values(by="Age", inplace=True)
print(df)

# Aggregation method
print("Statistics")
print("Total Salary:", df["Salary"].sum())
print("Average Salary:", df["Salary"].mean())
print("Maximum Salary:", df["Salary"].max())
print("Minimum Salary:", df["Salary"].min())
```

## Groupby Method

```python
import pandas as pd

data = {
    "Name": ["Rakesh", "Sourav", "Nitish", "Rohit", "Aarti"],
    "Department": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [50000, 45000, 52000, 48000, 51000]
}

df = pd.DataFrame(data)
print(df)

df2 = df.groupby("Department")["Salary"].sum()
print(df2)

# Multiple aggregation 
df3 = df.groupby("Department")["Salary"].agg(["sum", "mean", "max", "min"])
print(df3)
```
